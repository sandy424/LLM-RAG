```python
#라이브러리 임포트
from newspaper import Article
```

```python
#크롤링할 사이트(아무 뉴스 사이트 가능)
URL = "https://www.khan.co.kr/article/202510291105021"
```

```python
article = Article(URL, language='ko')
```

```python
import re
article.download()
article.parse()

context = article.text.split('\n')
context
```

여기서부터 문제


```python
#1.위의 문자열에서 ·, 콤마, (), “”, '' 문자열을 제거하는 코드를 작성하시오.
import re
removeword = re.sub(r'[·,()“”]|\'\'', '', str(context))
removeword
```

```python
#2.removeword를 토큰화 하시오 -> ([\w]+ 이용 하시오)
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w]+")

tokenizerd = tokenizer.tokenize(str(removeword))
print(tokenizerd)
```

```python
#3.Counter함수를 이용하여 토큰화한 단어들을 바탕으로 빈도수 순으로 내림차순 정렬하시오
from collections import Counter

token_freq = Counter(tokenizerd)

sorted_tokens = sorted(token_freq.items(), key=lambda x: x[1], reverse=True)

print(sorted_tokens)
```

```python
#4.위의 단어들을 딕셔너리 형태로 변환하고 단어의 수가 2이상인 단어들과 단어의 수를 출력하시오.
token_dict = dict(sorted_tokens)
filtered_tokens = [(token, freq) for token, freq in token_dict.items() if len(token)>=4]
print(filtered_tokens)
```

```python
#5.형태소 분석기Okt를 이용하여 명사만 남기고 나머지는 제거하는 코드를 작성하시오.
from konlpy.tag import Okt

okt = Okt()
pos_tags = okt.pos(" ".join(token_dict.keys()))
filtered_tokens = [(word, pos) for word,pos in pos_tags if pos in ['Noun']]
print(filtered_tokens)
```
